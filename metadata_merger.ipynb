### Merging image metadata with article metadata
Creating new sheet that still needs tropes + reprint info added

```python
import pandas as pd
```

```python

# 1. Load the two sheets
images_df = pd.read_csv("image_metadata_full.csv")     # object_id, article_id, image_display_template, ...
articles_df = pd.read_csv("atricle_metadata_starter.csv") # article_id, title, publication, ...

# 2. (Optional but recommended) drop image-related columns from article sheet
# since the image sheet already has the "real" image paths/templates
cols_to_drop = [
    "image_display_template",
    "image_parent_id",
    "image_file",
    "image_object_location",
    "image_file_small",
    "image_file_thumb",
]
articles_df = articles_df.drop(columns=cols_to_drop, errors="ignore")

# 3. Merge: LEFT JOIN so every image row keeps its place,
# and article metadata gets repeated for each matching article_id
merged = images_df.merge(
    articles_df,
    on="article_id",
    how="left"
)

# 4. Save to a new CSV
merged.to_csv("images_with_article_metadata.csv", index=False)
```
